# Census Data (2010 - 2019)

```python
# Dependencies
import pandas as pd
```

```python
# Create file path
file_path = "Resources/cc-est2019-alldata.csv"

# Read in csv as dataframe
# Note: I ran into an encoding error with this file that
# was resolved by setting the encoding as seen below.
# Not exactly sure why the error occurred though.
census_df = pd.read_csv(file_path, encoding = "ISO-8859-1")

# Display all rows 
pd.set_option('display.max_columns', None)

# View the dataframe
census_df.head(5)
```

## Data Processing

### Objectives:
- Filter Census Data for YEAR 2019. 
- Remove uninformative data.
- Convert data into a percentage of total population
- Adjust age / demographic information from 18 rows into 1 row
- Export processed data

```python
# Filter dataframe to only show census data from 2019 (which is #12)
# See cc-est2019-alldata.pdf for a key to interpreting the data
census_df = census_df.query("YEAR == 12")
census_df.head(5)
```

```python
# Drop "STATE", "COUNTY", and "YEAR" columns
census_df = census_df.drop(["STATE", "COUNTY", "YEAR"], axis=1)
census_df.head()
```

```python
# SUMLEV is "Geographic Summary Level"
census_df["SUMLEV"].unique()
```

```python
# Drop SUMLEV
census_df = census_df.drop(columns = ['SUMLEV'])
census_df.head()
```

```python
# Create a list of columns to convert to percentages
demo_list = census_df.columns.tolist()

# Remove unnecessary columns
del demo_list[0:3]

# Check that list is correct
demo_list
```

```python
# Loops through all columns and converts to a percentage
for column_name in demo_list:
    census_df["PERC_" + column_name] = (census_df[column_name] / census_df["TOT_POP"] * 100)
```

```python
# Loops through all columns and removes raw values
for column_name in demo_list:    
    census_df = census_df.drop([column_name], axis=1)

# Drop PERC_TOT_POP
census_df = census_df.drop(["PERC_TOT_POP"], axis=1)

# Checks that columns were removed
census_df.columns.tolist()
```

```python
# View the dataframe
census_df.head()
```

```python
census_df["State&County"] = census_df['STNAME'].astype(str) +", "+ census_df["CTYNAME"]
census_df = census_df.drop(["STNAME", "CTYNAME"], axis=1)
census_df.head()
```

```python
census_df.shape
```

```python
census_df.columns.tolist()
```

```python
# Rearrange the order of the columns to make State&County come first instead of last

# Add all columns to a list called cols
cols = census_df.columns.tolist()

# Rearrange the first and last item in the list
cols = cols[-1:] + cols[:-1]

# Add the new order to census_df
census_df = census_df[cols]

# View the dataframe
census_df.head()
```
