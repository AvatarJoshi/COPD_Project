```python
# Import dependencies
import pandas as pd
import geopandas
import geojson
```

```python
# Create dataframe with data for map
file_path = "./Resources/processed_PLACES_COPD.csv"
predicted_df = pd.read_csv(file_path)
predicted_df
```

```python
predicted_df = predicted_df.drop(["State_County"], axis=1)
predicted_df
```

```python
# Upload geoJSON file for map
file_path = "../COPD_Mapping/static/data/county.geojson"
county_df = geopandas.read_file(file_path)
county_df.head(25)
```

```python
county_df.rename(columns = {'NAME10':'County'}, inplace = True)
county_df.head()
```

```python
# Look at what numbers are used for State FIPS identifiers
county_df.STATEFP10.unique().tolist()
# It appears they are not simply 1-50. Need a key to translate these to the appropriate State name
```

```python
# Remove leading zeroes from STATEFP10
county_df['STATEFP10'] = county_df['STATEFP10'].str.lstrip('0')
county_df.head()
```

```python
# Check datatypes
county_df.dtypes
```

```python
# Get State FIPS key chart and make it a dataframe
file_path = "https://gist.githubusercontent.com/dantonnoriega/bf1acd2290e15b91e6710b6fd3be0a53/raw/11d15233327c8080c9646c7e1f23052659db251d/us-state-ansi-fips.csv"
fips_df = pd.read_csv(file_path)
fips_df.head()
```

```python
fips_df.columns = ["State", "STATEFP10", "State_Abbrv"]
fips_df.head()
```

```python
fips_df["STATEFP10"] = fips_df["STATEFP10"].astype('str')
```

```python
fips_df.dtypes
```

```python
# Merge dataframes
datamap_df = fips_df.merge(county_df, on="STATEFP10")
datamap_df.head(5)
```

```python
# Merge dataframes
datamap_df = predicted_df.merge(datamap_df, on=["State", "County"])
```

```python
# View the dataframe
pd.set_option('display.max_columns', 50)
datamap_df.head(5)
```

```python
# Remove unneeded data
datamap_df = datamap_df.drop(["State_Abbrv", "NAMELSAD10", "CSAFP10", "CBSAFP10", "METDIVFP10"], axis=1)
datamap_df.head()
```

```python
# Export file path
file_path = '../COPD_Mapping/static/data/data_map.geojson.zip'
```

```python
datamap_df.to_file(file_path, index=False, driver="GeoJSON")
```
